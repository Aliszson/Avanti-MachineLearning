
2. Faça a leitura de uma imagem de entrada 2D colorida qualquer e aplique o
filtro da média, mediana e gaussiano na imagem de entrada. Mostre os
resultados de cada filtro por meio da biblioteca matplotlib.

```python
from matplotlib import pyplot as plt

#Suavização com filtro da média
# Carregar a imagem
imagem_caminho = 'salt-pepper-noise.tif'
cinza_imagem = cv2.imread(imagem_caminho, 0)

# Aplicar filtro da média
imagem_borrada = cv2.blur(cinza_imagem, (5, 5)) # (5, 5) tamanho da janela

# Plotar a imagem cinza e imagem suavizada
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(cinza_imagem, cv2.COLOR_BGR2RGB))
plt.title('Imagem Cinza')
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(imagem_borrada, cv2.COLOR_BGR2RGB))
plt.title('Imagem Suavizada (Filtro de Média)')
plt.show()
```

```python
# Suavização com filtro gaussiano
imagem_borrada_gauss = cv2.GaussianBlur(cinza_imagem, (5,5), 1)

# Plotar a imagem cinza e suavizada
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(cinza_imagem, cv2.COLOR_BGR2RGB))
plt.title('Image Cinza')
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(imagem_borrada_gauss, cv2.COLOR_BGR2RGB))
plt.title('Imagem Suavizada (Filtro Gaussiano)')
plt.show()
```
